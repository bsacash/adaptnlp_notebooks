```
from math import sqrt

def cos_similarity(a,b):
  f = lambda x: x*x
  z = [*zip(a,b)]
  n = sum([*map(lambda x: x[0]*x[1], z)])
  d = sqrt(sum([*map(f, a)]))*sqrt(sum([*map(f, b)]))
  return n/d
```

##Stacked Embeddings

Stacked embeddings are a simple yet important concept pointed out by Flair that can help produce state-of-the-art results in downstream NLP models.

They are contextualized word embeddings like `EasyWordEmbeddings`, that are the concatenation of tensors from multiple language models. 

Let's look at an example where we produce a stacked word embedding from BERT base cased and XLNet base cased languagel models.

`EasyStackedEmbeddings` take in a variable number of key shortcut names to pre-trained language models.

##Document Embeddings

Document embeddings can load in a variable number of language models, just like in stacked embeddings, and produce and embedding for an entire text or document. Unlike `EasyWordEmbeddings` and `EasyStackedEmbeddings`, `EasyDocumentEmbeddings` will produce one contextualized embedding for a sequence of words using the pool or RNN method provided by Flair.
